## Introduction to Matplotlib

Matplotlib is a widely-used Python library for creating static, animated, and interactive visualizations. It provides a flexible and powerful interface for plotting a wide range of charts and graphs. Matplotlib integrates well with Jupyter notebooks and offers a variety of features to customize your plots.

In this notebook, we will cover the following topics:

1. **Basic Plotting**: Creating simple line plots and customizing their appearance.
2. **Adding Titles and Labels**: Enhancing the readability of your plots with titles and axis labels.
3. **Customization Options**: Exploring various options to modify plot styles, markers, and colors.
4. **Additional Plot Types**: Brief overview of other plot types available in Matplotlib.
5. **Advanced Features**: Customizing plots further with grids, legends, and multiple plots.


```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]
```

```python

# create a plot
plt.plot(x, y, marker='o', linestyle='-', color='b')

# adding titles and labels
plt.title('Simple Line Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.grid(True)
plt.show()
```

```python
plt.scatter(x, y, color='red', marker='x')
plt.title('Scatter Plot Example')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.grid(True)
plt.show()
```

```python
plt.plot(x, y, linewidth=2, linestyle='--', color='green')
plt.title('Customized Line Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.grid(True)
plt.show()
```

```python
plt.plot(x, y, marker='s', markersize=10, markeredgecolor='black', markerfacecolor='yellow')
plt.title('Customized Markers')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.grid(True)
plt.show()
```

Subplots and Axes

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# plot in the first subplot
axs[0, 0].plot(x, y, 'tab:blue')
axs[0, 0].set_title('Plot 1')

# plot in the second subplot
axs[0, 1].scatter(x, y, color='tab:orange')
axs[0, 1].set_title('Plot 2')

# plot in the third subplot
axs[1, 0].bar(x, y, color='tab:green')
axs[1, 0].set_title('Plot 3')

# plot in the fourth subplot
axs[1, 1].hist(y, bins=5, color='tab:red')
axs[1, 1].set_title('Plot 4')

plt.tight_layout()
plt.show()
```

Annotations and Text

```python
plt.plot(x, y, marker='o', linestyle='-', color='b')
plt.title('Plot with Annotations')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# adding annotations
for i, txt in enumerate(y):
    plt.annotate(txt, (x[i], y[i]), textcoords="offset points", xytext=(0,10), ha='center')

plt.grid(True)
plt.show()
```

Legends and Labels

```python
plt.plot(x, y, marker='o', linestyle='-', color='b', label='Data Line')
plt.title('Plot with Legend')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```

Saving and Exporting Figures

```python
plt.plot(x, y, marker='o', linestyle='-', color='b')
plt.title('Plot to Save')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.grid(True)
plt.savefig('plot.png', dpi=300)  # save as PNG with high resolution
```

Interactive Plots

For interactive plots, consider using %matplotlib notebook or %matplotlib widget magic commands

```python
%matplotlib notebook
plt.plot(x, y, marker='o', linestyle='-', color='b')
plt.title('Interactive Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.grid(True)
plt.show()
```

Customizing Plot Styles

```python
plt.style.use('ggplot')
plt.plot(x, y, marker='o', linestyle='-', color='b')
plt.title('Plot with ggplot Style')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.grid(True)
plt.show()
```
